```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

balaji = pd.read_csv('Balaji Fast Food Sales.csv')

balaji.info()
```

```python
balaji['date'] = balaji['date'].str.replace('/', '-')
balaji['date'] = pd.to_datetime(balaji['date'])
```

```python
balaji['weekday']=balaji['date'].dt.weekday + 1
balaji.head()
```

```python
day_order = ['Morning', 'Afternoon', 'Evening', 'Night', 'Midnight']

sales_per_TOD = balaji.groupby(['time_of_sale'], as_index = False)['transaction_amount'].count()

sales_per_TOD.rename(columns={'transaction_amount': 'total_sales'}, inplace=True) 

daily_time_counts = balaji.groupby(['time_of_sale'])['date'].nunique()

sales_per_TOD = pd.merge(sales_per_TOD, daily_time_counts, left_on='time_of_sale', right_index=True, how='outer')

sales_per_TOD.rename(columns={'date': 'total_TOD'}) 

avg_sales_per_TOD = sales_per_TOD['total_sales'] / sales_per_TOD['date']

avg_sales_per_TOD
# avg_sales_per_TOD.index = pd.Categorical(avg_sales_per_TOD.index, categories=day_order, ordered = True) 

# daily_time_counts

# avg_sales_per_TOD = avg_sales_per_TOD.map(lambda total_sales: ) 

# existing_dates
# sns.set_theme()
# ax = sns.lineplot(data=avg_sales_per_TOD, sort = day_order)
# ax.set(xlabel='Days', ylabel='Amount Spent')
# plt.title('Average Sales Per Time Of Day')

# plt.show()
```
